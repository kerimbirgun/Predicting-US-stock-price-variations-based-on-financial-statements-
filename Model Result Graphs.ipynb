```python
import pandas as pd
import numpy as np
from pmdarima.arima import auto_arima
from sklearn.metrics import r2_score
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('FW2 Model Results.csv')

fig = plt.figure(figsize=(15,8)) # Create matplotlib figure

df.set_index('Models Used', inplace=True)

ax = fig.add_subplot(111) # Create matplotlib axes
ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.

df['Adj R Squared'].plot(kind='line', #type of chart to use
                                color='orange', #colour of bar
                                ax=ax, #what matplotlinb axes to use
                                fontsize=15) #set font size 
df['Test RMSE'].plot(kind='line', #type of chart to use
                                color='brown', #colour of bar
                                ax=ax2, #what matplotlinb axes to use
                                fontsize=15) #set font size 

#Create ax customatisations
ax.set_ylabel('Adj R Squared',fontsize = 15) # name the primary y axis


#Create ax2 customatisations
ax2.set_ylabel('Test RMSE',fontsize = 15) #name the secondary y axis

#Global Plot settings
fig.legend(fontsize = 'xx-large') # set the legend
fig.suptitle('Model Results for R Squared & RMSE', fontsize=20)

#display the charts
plt.show() 
```

```python
df2 = pd.read_csv('FW2 Model RMSE Results.csv')

fig = plt.figure(figsize=(15,8)) # Create matplotlib figure

df2.set_index('Models Used', inplace=True)

ax = fig.add_subplot(111) # Create matplotlib axes
ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.

df2['Train RMSE'].plot(kind='line', #type of chart to use
                                color='orange', #colour of bar
                                ax=ax, #what matplotlinb axes to use
                                fontsize=15) #set font size 
df2['Test RMSE'].plot(kind='line', #type of chart to use
                                color='brown', #colour of bar
                                ax=ax2, #what matplotlinb axes to use
                                fontsize=15) #set font size 

#Create ax customatisations
ax.set_ylabel('Train RMSE',fontsize = 15) # name the primary y axis


#Create ax2 customatisations
ax2.set_ylabel('Test RMSE',fontsize = 15) #name the secondary y axis

#Global Plot settings
fig.legend(fontsize = 'xx-large') # set the legend
fig.suptitle('Model Results - Train vs Test RMSE', fontsize=20)

ax.set_xticklabels(ax.get_xticklabels(), #get a strings with the label names
                   rotation=45)

#display the charts
plt.show() 
```
